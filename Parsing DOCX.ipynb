```python
import docx
import re
import json
from collections import OrderedDict
import pandas as pd
import csv
```

## A) Senarai Klinik Perubatan Swasta Berdaftar Yang Boleh Menjalankan Saringan COVID-19 Dalam Kawasan Premis
- No contact
- No coordinate

```python
input_docx = r"DOCX\A) Senarai Klinik Perubatan Swasta Berdaftar Yang Boleh Menjalankan Saringan COVID-19 Dalam Kawasan Premis.docx"
output_csv = r"CSV\A) Senarai Klinik Perubatan Swasta Berdaftar Yang Boleh Menjalankan Saringan COVID-19 Dalam Kawasan Premis.csv"
```

```python
def clean_details(details, state):
    data = OrderedDict()
    data["Name"]=""
    data["Clinic"]=1
    data["Hospital"]=0
    data["ST"]=1
    data["VAC"]=0
    data["Latitude"]=""
    data["Longitude"]=""
    data["Address"]=""
    data["Contact"]=""
    data["State"]=""
    
    data["Name"] = details[1].text.replace("\t", "").strip().title()
    data["State"] = state

    data["Address"] = details[2].text.strip().title()
    return data
```

```python
doc = docx.Document(input_docx)
count = 1
data_ls = []
state = ""

for idx, table in enumerate(doc.tables):
    for row in table.rows:
        count += 1
        # print("\n".join([cell.text for cell in row.cells]))
        if row.cells[0].text.strip() == "BIL.":
            continue
        if row.cells[0].text != row.cells[1].text:
            data = clean_details(row.cells, state)
            data_ls.append(data)
        else:
            state = row.cells[0].text.strip().title()


print(count)
```

```python
df_1 = pd.DataFrame.from_dict(data_ls)
df_1.tail(n=10)
```

```python
# Extract duplicate rows
dup_df = df_1.loc[df_1.duplicated(subset=['Address']), :]
dup_df
```

```python
uni_df = df_1.drop_duplicates().reset_index()
uni_df
```

```python
uni_df.to_csv(
    output_csv,
    quoting=csv.QUOTE_NONNUMERIC,
    escapechar="\\",
    doublequote=False,
    index=False)
```
