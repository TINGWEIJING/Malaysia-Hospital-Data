```python
import json
import pandas as pd
from collections import OrderedDict
from geopy.geocoders import Nominatim
import csv
```

```python
locator = Nominatim(user_agent="myGeocoder")
```

## Klinik 1Malaysia.csv
- Missing Latitude & Longitude
- Addresses are not accurate and not very organised

```python
input_csv = r"Data\Klinik 1Malaysia.csv"
output_csv = r"CSV\Klinik 1Malaysia.csv"
```

```python
df = pd.read_csv(input_csv, header=0, index_col=0)
```

```python
df.head(n=1)
```

```python
def clean_details(details):
    data = OrderedDict()
    data["Name"]=""
    data["Clinic"]=1
    data["Hospital"]=0
    data["ST"]=0
    data["VAC"]=0
    data["Latitude"]=""
    data["Longitude"]=""
    data["Address"]=""
    data["Contact"]=""
    data["State"]=""
    
    data["Name"] = str(details.name).strip()
    data["State"] = details.state.strip().capitalize()
    data["Contact"] = details.tel.strip()

    raw_addr = [x.strip() for x in details.address.split(",")]
    addr = []

    if len(raw_addr) <= 2:
        addr.append(raw_addr[0])
        addr.append(raw_addr[1])
    elif len(raw_addr) <= 3:
        addr.append(", ".join(raw_addr[:2]))
        addr.append(raw_addr[2])
    else:
        addr.append(", ".join(raw_addr[:2]))
        addr.append(", ".join(raw_addr[2:]))

    addr.append(str(details.postcode))
    addr.append(data["State"])

    data["Address"] = "\n".join(addr)

    location = locator.geocode(f"{data['Name']} Malaysia")
    if location:
        data["Latitude"] = location.latitude
        data["Longitude"] = location.longitude

    return data
```

```python
data_ls = []
for row in df.itertuples(index=False):
    data = clean_details(row)
    data_ls.append(data)
```

```python
df_1 = pd.DataFrame.from_dict(data_ls)
df_1.head(n=1)
```

```python
df_1.to_csv(
    output_csv,
    quoting=csv.QUOTE_NONNUMERIC,
    escapechar="\\",
    doublequote=False,
    index=False)
```
